```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import StandardScaler,PolynomialFeatures,OneHotEncoder
from sklearn .metrics import r2_score
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
```

```python
data=pd.read_csv(r"C:\Users\mdfir\Downloads\Mental_Health_and_Social_Media_Balance_Dataset.csv")
```

```python
x=data.drop(columns=["Happiness_Index(1-10)"])
y=data["Happiness_Index(1-10)"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.2)
```

```python
# use of pipeline and columns transformations
cat=["Gender","Social_Media_Platform"]
num=["Age","Daily_Screen_Time(hrs)","Sleep_Quality(1-10)","Stress_Level(1-10)","Days_Without_Social_Media","Exercise_Frequency(week)"]
```

```python
cat_pipeline=Pipeline([

("encoder",OneHotEncoder(handle_unknown="ignore",sparse_output=False))])
```

```python
num_pipeline=Pipeline([
("scale",StandardScaler())])
```

```python
# ColumnTransformer use 
preprocess=ColumnTransformer([
("cat_pipeline",cat_pipeline,cat),
    ("num_pipeline",num_pipeline,num)    
])
```

```python
pip=Pipeline([
("preprocess",preprocess),
    ("model",LinearRegression())

    
])
```

```python
pip.fit(x_train,y_train)
```

```python
pred=pip.predict(x_test)
```

```python
pred
```

```python
score=r2_score(y_test,pred)
```

```python
score
```

```python
# SGDRegressor
model=Pipeline([
("preprocess",preprocess),
    ("model",SGDRegressor(alpha=0.2,eta0=0.01,learning_rate="constant",max_iter=300))

    
])
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
score=r2_score(y_test,pred)
```
